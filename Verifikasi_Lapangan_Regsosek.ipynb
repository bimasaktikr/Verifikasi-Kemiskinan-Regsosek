# Filter 1 - Tanpa Anak KOS tanpa ART PNS

```python
rt_join_art = pd.merge(left=df_rt, right=df_art, left_on="id_rt_new", right_on="id_rt_new")
```

```python
rt_join_art.shape
```

```python

# df_tanpakos = rt_join_art.loc[~((rt_join_art['r112']=='1') &
#                           (rt_join_art['r301a']=='2') &
#                           (rt_join_art['r412']=='2'))]
                          
id_rt_new_kos = rt_join_art.loc[((rt_join_art['r112']=='1') &
                          (rt_join_art['r301a']=='2') &
                          (rt_join_art['r412']=='2')), "id_rt_new"]


# id_rt_new_kos  
```

```python
id_rt_new_PNS = rt_join_art.loc[rt_join_art['r418'] == '5', "id_rt_new"]
# id_rt_new_PNS
```

```python
# KRT TIDAK TINGGAL DIRUMAH

df_rt_new_umurKRT_nan = rt_join_art.loc[(rt_join_art['r401'] == '1') & (rt_join_art['r407'].isna()), "id_rt_new"]     
# df_rt_new_umurKRT_nan
```

```python
filter = [id_rt_new_kos  , id_rt_new_PNS, df_rt_new_umurKRT_nan]

filter_concat = pd.concat(filter)

# filter_concat
```

```python
df_rt_filter1 = df_rt.loc[~(df_rt['id_rt_new'].isin(filter_concat))]
# df_rt_filter1
```

**Filter 9 Variabel**

```python
df_final = df_rt_filter1

#check apakah ada data NaN di R112 atau R302
df_final_filter_Nan = df_final.loc[df_final['r112'].isna() ]
df_final_filter_Nan = df_final_filter_Nan.loc[df_final_filter_Nan['r302'].isna() ]

#remove semua data dengan NaN Value di R112 atau R302
df_final_dropNA = df_final.dropna(subset=['r112', 'r302'])



df_final_dropNA.shape
# df_final ['v1'] = df_final.apply(lambda row: 1 if int(row['r302']) / int(row['r112']) < 7 else 0, axis=1)
```

Filter Variabel 1 - Rumah Tangga Dengan Luas Perkapita dibawah 7m2

```python
# df_final_dropNA

df_final_dropNA['v1'] = df_final_dropNA.apply(lambda row: 1 if ((int(row['r302']) / int(row['r112']) < 7 ) and (row['r301a'] != 4)) else 0, axis=1)

df_final_dropNA_v1_True = df_final_dropNA.loc[df_final_dropNA['v1'] == 1]
df_final_dropNA_v1_True
```

Jenis Lantai Terluas

```python
df_final_dropNA.shape

df_final_dropNA['v2'] = df_final_dropNA.apply(lambda row: 1 if ((int(row['r302']) / int(row['r112']) < 7 ) and (row['r303'] != 1)) else 0, axis=1)

df_final_dropNA_v2_True = df_final_dropNA.loc[df_final_dropNA['v2'] == 1]
df_final_dropNA_v2_True
```

Jenis Dinding Terluas

```python
# df_final_dropNA

df_final_dropNA['v3'] = df_final_dropNA.apply(lambda row: 1 if (((int(row['r302']) / int(row['r112']) < 7 ) and (row['r304'] == 1)) or (row['r304']!= "1")) else 0, axis=1)

df_final_dropNA_v3_True = df_final_dropNA.loc[df_final_dropNA['v3'] == 1]
df_final_dropNA_v3_True
```

Variabel 4 - Jenis Atap Terluas 
Selain Beton dengan LPK 6m

```python
# df_final_dropNA

df_final_dropNA['v4'] = df_final_dropNA.apply(lambda row: 1 if ((int(row['r302']) / int(row['r112']) < 7 ) and (row['r305'] != 1)) else 0, axis=1)

df_final_dropNA_v4_True = df_final_dropNA.loc[df_final_dropNA['v4'] == 1]
df_final_dropNA_v4_True
```

Variabel 5 - Sumber Air Minum Utama 
LPK < 7 -> Air Kemasan, Aor Isi Ulang & Leding
dan semua kode 4,5,6,7,8,9,10,11

```python
# df_final_dropNA
v5 = ['1', '2', '3']

df_final_dropNA['v5'] = df_final_dropNA.apply(lambda row: 1 if (((int(row['r302']) / int(row['r112']) < 7 ) and (row['r306a'] in v5 )) or (row['r306a'] not in v5 )) else 0, axis=1)

df_final_dropNA_v5_True = df_final_dropNA.loc[df_final_dropNA['v5'] == 1]
df_final_dropNA_v5_True
```

Variabel 6 - Daya Listrik 

LPK < 7 => bukan listri, 450w atau 900 w 

```python
# df_final_dropNA
v6 = ['1', '2', '4']

df_final_dropNA['v6'] = df_final_dropNA.apply(lambda row: 1 if (((int(row['r302']) / int(row['r112']) < 7 ) and (row['r307a'] in v6 )) and (row['r307b1'] in ['1', '2'] )) else 0, axis=1)

df_final_dropNA_v6_True = df_final_dropNA.loc[df_final_dropNA['v6'] == 1]
df_final_dropNA_v6_True
```

Variabel 7 - Bahan bakar Memasak 

LPK < 7 => 308 [1,4,7,,9,10,00] 

```python
# df_final_dropNA
v7 = ['1', '4', '7', '9', '10', '00', '0', '11']

df_final_dropNA['v7'] = df_final_dropNA.apply(lambda row: 1 if ((int(row['r302']) / int(row['r112']) < 7 ) and (row['r308'] in v7 ))  else 0, axis=1)

df_final_dropNA_v7_True = df_final_dropNA.loc[df_final_dropNA['v7'] == 1]
df_final_dropNA_v7_True
```

Variabel 8 - Fasilitas BAB

LPK < 7 => memiliki sendiri OR tidak memiliki

```python
# df_final_dropNA

df_final_dropNA['v8'] = df_final_dropNA.apply(lambda row: 1 if (((int(row['r302']) / int(row['r112']) < 7 ) and (row['r309a'] in ['1','2'])) or (row['r309a'] not in [ '1', '2']))  else 0, axis=1)

df_final_dropNA_v8_True = df_final_dropNA.loc[df_final_dropNA['v8'] == 1]
df_final_dropNA_v8_True
```

Variabel 9 - Kepemilikan Aset

keluarkan Mobil, Kapal, AC, Water Heater

```python
# df_final_dropNA['r502c']

df_final_dropNA['v9'] = df_final_dropNA.apply(lambda row: 1 if (row['r502c'] == '2' or row['r502d'] == '2' or row['r502k'] == '2' or row['r502m'] == '2')  else 0, axis=1)

df_final_dropNA_v9_True = df_final_dropNA.loc[df_final_dropNA['v9'] == 1]
df_final_dropNA_v9_True
```

Variabel 10 - Umur KRT
Laki-Laki >50 perempuan >45

```python
#join ART dengan RT hasil Filter
rt_join_art_filtered = pd.merge(left=df_final_dropNA, right=df_art, left_on="id_rt_new", right_on="id_rt_new")
rt_join_art_filtered
```

```python
rt_join_art_filtered['v10'] = rt_join_art_filtered.apply(lambda row: 1 if ((row['r401'] == '1' and row['r405'] == '1' and int(row['r407']) >= 50) or (row['r401'] == '1' and row['r405'] == '2' and int(row['r407']) >= 45))  else 0, axis=1)

rt_join_art_filtered_v10_True = rt_join_art_filtered.loc[rt_join_art_filtered['v10'] == 1]
rt_join_art_filtered_v10_True.shape

id_rt_new_umurKRT = rt_join_art_filtered_v10_True['id_rt_new']
id_rt_new_umurKRT
```

```python
umurkrt = list(id_rt_new_umurKRT)
umurkrt
```

```python
pendidikan_tinggi = ['18', '19', '20', '21', '22']
rt_join_art_filtered['v11'] = rt_join_art_filtered.apply(lambda row: 1 if (row['r415'] not in pendidikan_tinggi )  else 0, axis=1)

rt_join_art_filtered_v11_True = rt_join_art_filtered.loc[rt_join_art_filtered['v11'] == 1]
id_rt_new_pendidikanKRT = rt_join_art_filtered_v11_True['id_rt_new']
id_rt_new_pendidikanKRT = id_rt_new_pendidikanKRT.drop_duplicates()
id_rt_new_pendidikanKRT
```

```python
#membuat variabel v10 di dataframe utama
df_final_dropNA['id_rt_new']
df_final_dropNA['v10'] = df_final_dropNA.apply(lambda row: 1 if (row['id_rt_new'] in umurkrt )  else 0, axis=1)

df_final_dropNA_v10_True = df_final_dropNA.loc[df_final_dropNA['v10'] == 1]
df_final_dropNA_v10_True
```

```python
#membuat variabel v11 di dataframe utama
pendidikanKrt = list(id_rt_new_pendidikanKRT)

df_final_dropNA['v11'] = df_final_dropNA.apply(lambda row: 1 if (row['id_rt_new'] in pendidikanKrt )  else 0, axis=1)

df_final_dropNA_v11_True = df_final_dropNA.loc[df_final_dropNA['v11'] == 1]
df_final_dropNA_v11_True
```
